```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Valor de gama, o coeficiente de expanso adiabtica. Ele exprime a razo entre a capacidade trmica
# a presso constante (Cp) e a capacidade trmica a volume constante (Cv), para um gs ideal.
# gamma = Cp/Cv = 1.4 para o ar a 20C (gs diatmico ideal) .
gamma = 1.4
# Vamos escolher trs valores diferentes de volume de gs para calcular as trajetrias dos processos.
V1 = V4 = 16.0
V2 = 12.0
V3 = 12.0

# Definio da temperatura nas quatro fases do ciclo
T1 = 60
T2 = 104
T3 = 220
T4 = 145 

# Vamos usar a equao de estado dos gases ideais: PV = nRT, com nR = 1, por simplicidade. Logo: PV = T.
#....Clculo de Presso nas quatro fases do ciclo 
P1 = T1 / V1  # Presso no Ponto 1 do Diagrama PV.
P2 = T2 / V2  # Presso no Ponto 1 do Diagrama PV.
P3 = T3 / V3  # Presso no Ponto 1 do Diagrama PV.
P4 = T4 / V4  # Presso no Ponto 1 do Diagrama PV.

plt.figure(figsize = (10, 6))

##################....Ponto 1....##################
# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
V_comp_adiabatica = np.linspace(V1, V2, 100)
P_comp_adiabatica = P1 * ((V1 / V_comp_adiabatica) ** gamma) # 
plt.plot(V_comp_adiabatica, P_comp_adiabatica, 'b', label = 'Compresso Adiabtica')

# ##################....Ponto 2....##################
# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
# V_comp_isocorica = np.linspace(V4, V1, 100)
# P_comp_isocorica = np.linspace(P4, T1 / V1, 100)
V_comp_isocorica = np.linspace(V3, V2, 100)
P_comp_isocorica = np.linspace(P3, T2 / V2, 100)
plt.plot(V_comp_isocorica, P_comp_isocorica , 'r', label = 'Compresso Isocrica')

##################....Ponto 3....##################
# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
V_exp_isentropica = np.linspace(V3, V4, 100)
P_exp_isentropica = P3 * ((V3 / V_exp_isentropica) ** gamma) # 
plt.plot(V_exp_isentropica, P_exp_isentropica, 'g', label = 'Expanso Isentrpica')

# ##################....Ponto 4....##################
# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
P_descomp_isocorica = np.linspace(P4, P1, 100)
V_descomp_isocorica = np.linspace(V4, V1, 100)
# V_comp_isocorica = np.linspace(V4, V1, 100)
# P_comp_isocorica = np.linspace(P4, T3 / V3, 100)
plt.plot(V_descomp_isocorica, P_descomp_isocorica, 'orange', label = 'Descompresso Isocrica')

##################....Plotagem do Grfico....##################
# Plotagem do diagrama PV.

plt.xlabel('Volume (U.A.)')
plt.ylabel('Presso (U.A.)')
plt.title('Ciclo Otto no Diagrama PV')
plt.legend()
plt.grid(True)

# Setas indicando os incios dos processos do ciclo.
plt.annotate('1', xy = (V1, P1), xytext = (V1 - 0.05, P1 + 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('2', xy = (V2, P2), xytext = (V2 + 0.05, P2 - 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('3', xy = (V3, P3), xytext = (V3 + 0.05, P3 - 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('4', xy = (V4, P4), xytext = (V4 - 0.05, P4 + 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Valor de gama, o coeficiente de expanso adiabtica. Ele exprime a razo entre a capacidade trmica
# a presso constante (Cp) e a capacidade trmica a volume constante (Cv), para um gs ideal.
# gamma = Cp/Cv = 1.4 para o ar a 20C (gs diatmico ideal) .
gamma = 1.4 

# Vamos escolher trs valores diferentes de volume de gs para calcular as trajetrias dos processos.
V1 = V4 = 15.0
V2 = V3 = 8.0

# O ponto 1 = (P1, V1) define o incio do ciclo com a expanso isobrica do gs. Vamos usar a equao de estado dos gases ideais: PV = nRT, com nR = 1, por simplicidade. Logo: PV = T.
T1 = 20  # Temperatura no Ponto 1 do Diagrama PV. T1 = 2.0
P1 = T1 / V1  # Presso no Ponto 1 do Diagrama PV.

# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
V_comp_adiabatica = np.linspace(V4, V1, 100)
P_comp_adiabatica = P1 * ((V1 / V_comp_adiabatica) ** gamma) # Dvida &&&&&&&&&&&


# Temperatura no Ponto 2 do Diagrama PV  a mesma que a do Ponto 1. Essa temperatura deve ser a mesma que aquela do Ponto 1, pois ocorreu uma expanso isotrmica.
T2 = T1
# Gerando os pontos para a curva que representa a expanso isotrmica do gs.
V_exp_isotermica = np.linspace(V1, V2, 100)
P_exp_isotermica = T1 / V_exp_isotermica



T2 = (V2 / V1) * T1      
# Gerando os pontos para a curva que representa a expanso isobrica do gs.
P2 = T2/V2
V_exp_isobarica = np.linspace(V1, V2, 100)
P_exp_isobarica = np.linspace(P1, P2, 100)


# Ponto 2: Incio do segundo processo - expanso adiabtica do gs. 
# No Ponto 2 do ciclo, o volume do gs  V2 e a presso  P2:
# Ponto 3: Final do segundo processo - expanso adiabtica do gs.
# Para um processo adiabtico, P*V^gamma = constante, ento P3*V3^gamma = P2*V2^gamma. 
# Podemos encontrar o valor da presso P3 usando os valores dos volumes V2 e V3, conhecendo P2.
P3 = P2 * (V2 / V3) ** gamma
# A temperatura no Ponto 3  dada pela relao T*V^(gamma-1) = constante (expanso adiabtica):
T3 = T2 * (V2 / V3)**(gamma-1)
# Gerando os pontos para a curva que representa a expanso adiabtica do gs.
V_exp_adiabatica = np.linspace(V2, V3, 100)
P_exp_adiabatica = P2 * ((V2 / V_exp_adiabatica) ** gamma)

# O ponto 3 = (P3, V2) define o incio do processo de compresso adiabtica do gs.
P3 = P2 * ((V2 / V3) ** gamma)

# Gerando os pontos para a curva que representa a compresso adiabtica do gs.
V_comp_adiabatica_2 = np.linspace(V2, V4, 100)
P_comp_adiabatica_2 = P3 * ((V3 / V_comp_adiabatica_2) ** gamma)

# Temperatura no Ponto 4 do Diagrama PV  a mesma que a do Ponto 1. Essa temperatura deve ser a mesma que aquela do Ponto 1, pois ocorreu uma compresso isotrmica.
T4 = T1
# Gerando os pontos para a curva que representa a compresso isotrmica do gs.
V_comp_isotermica = np.linspace(V3, V4, 100)
P_comp_isotermica = T1 / V_comp_isotermica

# Plotagem do diagrama PV.
plt.figure(figsize = (10, 6))
plt.plot(V_exp_isobarica, P_exp_isobarica, 'b', label = 'Expanso Isobrica')
plt.plot(V_exp_adiabatica, P_exp_adiabatica, 'r', label = 'Expanso Adiabtica')
plt.plot(V_descomp_isocorica, P_descomp_isocorica, 'g', label = 'Compresso Isotrmica')
plt.plot(V_comp_adiabatica, P_comp_adiabatica, 'orange', label = 'Compresso Adiabtica')
plt.xlabel('Volume (U.A.)')
plt.ylabel('Presso (U.A.)')
plt.title('Ciclo Otto no Diagrama PV')
plt.legend()
plt.grid(True)


# Setas indicando os incios dos processos do ciclo.
plt.annotate('1', xy = (V1, P1), xytext = (V1 - 0.05, P1 + 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('2', xy = (V2, P2), xytext = (V2 + 0.05, P2 - 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('3', xy = (V3, P3), xytext = (V3 + 0.05, P3 - 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.annotate('4', xy = (V4, P4), xytext = (V4 - 0.05, P4 + 0.05), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.show()
```
